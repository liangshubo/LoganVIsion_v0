```python
import os
import shutil


def move_subfolder(parent_dir: str, target_dir: str) -> None:
    """
    将 parent_dir 中的子文件夹整体移动到 target_dir 中
    """

    # 1. 确保目标目录存在
    os.makedirs(target_dir, exist_ok=True)

    # 2. 找到 parent_dir 下的所有子文件夹
    subfolders = [
        name for name in os.listdir(parent_dir)
        if os.path.isdir(os.path.join(parent_dir, name))
    ]

    # 3. 判断是否存在子文件夹
    if len(subfolders) == 0:
        print("没有找到子文件夹")
        return

    # 4. 逐个移动子文件夹
    for subfolder_name in subfolders:
        source_path = os.path.join(parent_dir, subfolder_name)
        target_path = os.path.join(target_dir, subfolder_name)

        print(f"找到子文件夹：{subfolder_name}")

        # 如果目标位置已经有同名文件夹，避免直接覆盖
        if os.path.exists(target_path):
            print(f"目标目录已存在同名文件夹，跳过：{target_path}")
            continue

        shutil.move(source_path, target_path)
        print(f"已移动到：{target_path}")


if __name__ == "__main__":
    parent_dir = r"/home/liangshubo/Project/CTSR/dataset/rawdata/LDIC/LDIC_SLICE/HR_all"      # 你的父文件夹
    target_dir = r"/home/liangshubo/Project/CTSR/dataset/rawdata/LDIC/LDIC_SLICE"      # 你要移动到的指定目录

    move_subfolder(parent_dir, target_dir)
```
